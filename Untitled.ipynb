```python
# Memilih fitur (features) dan target
features = ['age', 'sex', 'cp',	'trestbps', 'chol',	'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
X = df[features]
y = df['target']
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score,precision_recall_curve

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Perbaikan model (hyperparamater tuning)

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=5,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Akurasi:", accuracy)
print("Presisi:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Confusion Matrix:\n", cm)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ambil nilai importance
importance = model.feature_importances_

# Buat DataFrame dan urutkan
fi_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

# Normalisasi importance untuk gradasi warna
norm = (fi_df['Importance'] - fi_df['Importance'].min()) / (fi_df['Importance'].max() - fi_df['Importance'].min())

# Buat colormap pink (mirip heatmap yang kamu pakai)
colors = plt.cm.Reds(norm)  # Reds  pink/merah gradasi

plt.figure(figsize=(12, 6))
plt.bar(fi_df['Feature'], fi_df['Importance'], color=colors)
plt.xlabel("Feature")
plt.ylabel("Feature Importance")
plt.title("Visualisasi Feature Importance Pada Model Desicion Tree")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("grfik_fda.png")
plt.show()
```
